```python
import turtle
import math
s = turtle.Screen ()
s.setup (1000,700)
turtle.title ("Ngôi Nhà")
turtle.bgcolor ("#DCDCDC")

t = turtle.Turtle ()

# vẽ ngôi nhà
# điểm bắt đầu
x0 = -100
y0 = -100

# kich thuoc nha
crnha = 250
ccnha = 150
crmai = 180
ccmai = 60
ccongkhoi = 15
cccua = 2 / 3 * ccnha
crcua = crnha / 4

t.penup ()
t.goto (x0, y0)
t.pendown ()

t.fd (crnha)
t.rt (90)
t.fd (ccnha)
t.rt (90)
t.fd (crnha)
t.rt (90)
t.fd (ccnha)
t.rt (90)

t.goto (x0 + crmai / 2, y0 + ccmai)
t.goto (x0 + crmai, y0)
t.goto (x0, y0)

# vẽ ống khói
# pt duong cheo mai nha: y = (-2ccmai/crmai)x + 2ccmai
t.penup ()
t.goto (x0 + 2 / 3 * crmai, y0 + 2 / 3 * ccmai)
t.pendown ()
t.lt (90)
t.fd (ccongkhoi)
t.rt (90)
t.fd (crmai / 6)
t.goto (x0 + 5 / 6 * crmai, y0 + 1 / 3 * ccmai)

t.penup ()
t.goto (x0 + 2 / 3 * crmai + 3, y0 + 2 / 3 * ccmai + ccongkhoi + 5)
t.pendown ()
t.goto (x0 + 2 / 3 * crmai + 3, y0 + 2 / 3 * ccmai + ccongkhoi + 15)

t.penup ()
t.goto (x0 + 2 / 3 * crmai + 10, y0 + 2 / 3 * ccmai + ccongkhoi + 5)
t.pendown ()
t.goto (x0 + 2 / 3 * crmai + 10, y0 + 2 / 3 * ccmai + ccongkhoi + 25)

# vẽ cửa
t.penup ()
t.goto (x0 + 1 / 3 * crnha, y0 - ccnha)
t.pendown ()
t.lt (90)
t.fd (cccua)
t.rt (90)
t.fd (crcua)
t.goto (x0 + 1 / 3 * crnha + crcua, y0 - ccnha)

# vẽ cây
# toạ độ giữa cây 1
x1 = 300
y1 = -250
cccay = 210
crcay = 120
t.penup ()
t.goto (x1, y1)
t.pendown ()
t.fd (10)
t.rt (80)
t.fd (cccay / 3)
t.lt (80)
# 80o = 1.396 rad
t.bk (20 + 2 * cccay / 3 * math.cos (1.396))
t.lt (80)
t.fd (cccay / 3)
t.rt (80)
t.fd (10)

t.fd (crcay / 2)
t.goto (x1, y1 + cccay / 3)
t.goto (x1 - crcay / 2, y1)
t.goto (x1, y1)

t.penup ()
t.goto (x1 + crcay / 2 - 10, y1 + cccay / 3)
t.pendown ()
t.goto (x1 , y1 + cccay * 5 / 9)
t.goto (x1 - (crcay / 2 - 10), y1 + cccay / 3)
t.goto (x1 + crcay / 2 - 10, y1 + cccay / 3)

t.penup ()
t.goto (x1 + crcay / 2 - 20, y1 + cccay * 5 / 9)
t.pendown ()
t.goto (x1 , y1 + cccay * 7 / 9)
t.goto (x1 - (crcay / 2 - 20), y1 + cccay * 5 / 9)
t.goto (x1 + crcay / 2 - 20, y1 + cccay * 5 / 9)


# vẽ cây 2
# toạ độ giữa cây 1
x2 = -320
y2 = -100
cccay = 100
crcay = 50
t.penup ()
t.goto (x2, y2)
t.pendown ()
t.fd (10)
t.rt (80)
t.fd (cccay / 3)
t.lt (80)
# 80o = 1.396 rad
t.bk (20 + 2 * cccay / 3 * math.cos (1.396))
t.lt (80)
t.fd (cccay / 3)
t.rt (80)
t.fd (10)

t.fd (crcay / 2)
t.goto (x2, y2 + cccay / 3)
t.goto (x2 - crcay / 2, y2)
t.goto (x2, y2)

t.penup ()
t.goto (x2 + crcay / 2 - 10, y2 + cccay / 3)
t.pendown ()
t.goto (x2 , y2 + cccay * 5 / 9)
t.goto (x2 - (crcay / 2 - 10), y2 + cccay / 3)
t.goto (x2 + crcay / 2 - 10, y2 + cccay / 3)

t.penup ()
t.goto (x2 + crcay / 2 - 20, y2 + cccay * 5 / 9)
t.pendown ()
t.goto (x2 , y2 + cccay * 6 / 9)
t.goto (x2 - (crcay / 2 - 20), y2 + cccay * 5 / 9)
t.goto (x2 + crcay / 2 - 20, y2 + cccay * 5 / 9)

t.hideturtle ()

turtle.done ()
```
